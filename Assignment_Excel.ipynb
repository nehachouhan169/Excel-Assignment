### 1) Explain the difference between Absolute, Relative, and Mixed Cell Referencing in Excel with examples.
#### **Absolute Cell Referencing** : Fixed reference — does not change.When you copy a formula, both row and column stay the same.
##### **Example**
- Formula in C1 : =A1*$B$1
- Copy to C2 → formula remains : =A2*$B$1
#### **Use case** : When you want to use a constant value (tax rate, discount, fixed multiplier).
#### **Relative Cell Referencing** : Default type of reference in Excel.When you copy a formula, both the row and column change based on the new position.
##### **Example**
- Formula in C1:=A1+B1-
- Copy the formula to C2 → it becomes : =A2+B2
#### **Use case** : When you want the formula to adjust automatically for each row or column (like totals, calculations for many rows).
#### **Mixed Cell Referencing** : Partially fixed reference.Either row is fixed or column is fixed, but not both.
##### **Example**
- Formula : =$A1*B$1
- When copied : Column A stays fixed , Row 1 stays fixed
#### **Use case** : Very useful in tables, multiplication grids, mark sheets, etc.

### 2) What is a Macro in Excel? How does it help in automation?
#### > A Macro in Excel is a recorded or written set of instructions that tells Excel to perform a series of tasks automatically.Macros are written in VBA (Visual Basic for Applications).A macro is a shortcut that does many steps in one click.
#### > Working of Macro : Excel records your actions (like clicking, typing, formatting) and saves them as a program.Later, when you run the macro, Excel repeats the same actions automatically.
**How Macros Help in Automation**
- Saving time - repetitive tasks are done instantly
- Reducing errors - avoids manual mistakes
- Improving productivity - one-click execution
- Standardizing work - same format & process every time

### 3) What are Text Functions in Excel? Mention any five with examples.
#### > Text Functions in Excel are built-in functions used to manipulate, combine, extract, or format text (strings) in cells.They help in cleaning data, formatting names, and handling text-based information easily.
#### **Five Text Functions with Examples**
##### **LEFT()** : Extracts characters from the left side of text.
##### Syntax : LEFT(text, num_chars)
##### Example : =LEFT("EXCEL", 2)
###### Output : EX
#### **RIGHT()** : Extracts characters from the right side of text.
##### Syntax : RIGHT(text, num_chars)
##### Example : =RIGHT("EXCEL", 3)
##### Output : CEL
#### **MID()** : Extracts text from the middle of a string.
##### Syntax : MID(text, start_num, num_chars)
##### Example : =MID("EXCEL", 2, 3)
##### Output : XCE
#### **UPPER()** : Converts text to uppercase letters.
##### Syntax : UPPER(text)
##### Example : =UPPER("excel")
##### Output : EXCEL
#### **LOWER()** : Converts text to lowercase letters.
##### Syntax : LOWER(text)
##### Example : =LOWER("EXCEL")
##### Output : excel

### 5) Define the purpose of VLOOKUP and HLOOKUP. How are they different from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?
#### **VLOOKUP (Vertical Lookup)**
##### **Purpose** : Used to search for a value in the first column of a table and return a corresponding value from another column vertically.
##### Syntax : VLOOKUP(lookup_value, table_array, col_index, [range_lookup])
#### **HLOOKUP (Horizontal Lookup)**
##### **Purpose** : Used to search for a value in the first row of a table and return a corresponding value from another row horizontally.
##### Syntax : HLOOKUP(lookup_value, table_array, row_index, [range_lookup])
#### **How they are differnt from XLOOKUP** : VLOOKUP and HLOOKUP can search data only in one direction and require the lookup value to be in the first column or row. They may give incorrect results if columns or rows are added or removed. XLOOKUP is more flexible because it can search in any direction, works even if the lookup column is not first, and handles errors automatically.
#### **Best choice between XLOOKUP and INDEX-MATCH are** :
#### > XLOOKUP is best for most users because it is easier to write, easier to understand, and less error-prone. It works in any direction, handles errors automatically, and does not break when columns are inserted or deleted.
#### > INDEX-MATCH is better only in special cases, such as working with very large datasets or older Excel versions where XLOOKUP is not available.
#### > Conclusion:
- For regular and practical usage → XLOOKUP
- For advanced or compatibility needs → INDEX-MATCH
